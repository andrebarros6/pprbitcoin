```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
btc_eur = pd.read_csv(r"C:\Users\abarr\Desktop\ironhack\extra\pprbitcoin\pprbitcoin\BTC_EUR Kraken Historical Data (1).csv")
btc_eur.head()
```

```python
btc_eur.dtypes
```

```python
btc_eur['Date'] = pd.to_datetime(btc_eur['Date'])
```

```python
btc_eur['year'] =btc_eur['Date'].dt.year
```

```python
btc_eur.head()
```

```python
btc_eur['Price'] = btc_eur['Price'].str.replace(',','')
btc_eur['Price'] = btc_eur['Price'].astype(float)
```

```python
btc_eur.dtypes
```

```python
imob_pt = pd.read_csv(r"C:\Users\abarr\Desktop\ironhack\extra\pprbitcoin\pprbitcoin\m2-casas-PT.csv")
imob_pt.head()
```

```python
imob_pt['Date'] = pd.to_datetime(imob_pt['Mes'], format='%d-%m-%Y')
```

```python
imob_pt.head()
```

```python
imob_pt['year'] = imob_pt['Date'].dt.year
```

```python
imob_pt['month'] = imob_pt['Date'].dt.month
```

```python
imob_pt.head(9)
```

```python
imob_pt['Variacao mensal'] = imob_pt['Variacao mensal'].str.replace('%','')
imob_pt['Variacao mensal'] = imob_pt['Variacao mensal'].str.replace(',','.')
imob_pt['Variacao mensal'] = imob_pt['Variacao mensal'].astype(float)
```

```python
imob_pt['Variacao trimestral'] = imob_pt['Variacao trimestral'].str.replace('%','')
imob_pt['Variacao trimestral'] = imob_pt['Variacao trimestral'].str.replace(',','.')
imob_pt['Variacao trimestral'] = imob_pt['Variacao trimestral'].astype(float)
```

```python
imob_pt['Variacao anual'] = imob_pt['Variacao anual'].str.replace('%','')
imob_pt['Variacao anual'] = imob_pt['Variacao anual'].str.replace(',','.')
imob_pt['Variacao anual'] = imob_pt['Variacao anual'].astype(float)
```

```python
imob_pt.dtypes
```

```python
imob_pt.rename(columns={'Variacao mensal':'Variacao mensal [%]', 'Variacao trimestral':'Variacao trimestral [%]', 'Variacao anual':'Variacao anual [%]'}, inplace=True)
```

```python
imob_pt.head()
```

```python
df = btc_eur.merge(imob_pt[['Date','Preco m2 [EUR]']], on='Date', how='right')

df.columns
```

```python
df.head(10)
```

```python
df.drop(columns=(['Open','High','Low','Change %','Vol.']), inplace=True)
```

```python
df
```

```python
df['Preco m2 [BTC]'] =   df['Preco m2 [EUR]']/df['Price']
df
```

```python
df = df.sort_values('Date', ascending=True).reset_index()
```

```python
df.head()
```

```python
# Create the plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot house prices on left y-axis
color1 = 'tab:blue'
ax1.set_xlabel('Date')


ax1.set_ylabel('Preço por m² (EUR)', color='green')
ax1.plot(df['Date'], df['Preco m2 [EUR]'], color='green', marker='o', label='Preço/m² [EUR]')
ax1.tick_params(axis='y', labelcolor='green')
ax1.grid(True, alpha=0.3)

# Create second y-axis for Bitcoin price
ax2 = ax1.twinx()
color2 = 'tab:orange'
ax2.set_ylabel('Preço por m² (BTC)', color=color2)

ax2.set_yscale('log')
ax2.yaxis.set_major_formatter(ticker.ScalarFormatter(useMathText=False))

ax2.plot(df['Date'], df['Preco m2 [BTC]'], color=color2, marker='s', label='Preço/m² [BTC]')
ax2.tick_params(axis='y', labelcolor=color2)

# Add title and legends
plt.title('Preço por m² em Portugal - EUR vs BTC', fontsize=14, fontweight='bold')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
